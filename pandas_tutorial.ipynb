**Selecionando colunas com condições especificas**

**Para criar condições utilizamos:**

Use & | ~ (not and, or, not)

Enclose each conditional expression in parentheses when using comparison operators

```python
# Aqui criamos um dataframe com as informações do tipo1 igual ao estilo fogo, que é o estilo principal
Tipo_fogo = poke.loc[poke['Type 1']=="Fire"]
Tipo_fogo


# Aqui criamos um dataframe com duas condições, sendo elas o tipo 1 dark E tipo 2 fire, ou seja, 
# Apenas quando as duas forem true
Tipo_fogo_dark = poke.loc[(poke['Type 1']=="Dark") & (poke["Type 2"]=="Fire")]
Tipo_fogo_dark

# Aqui criamos um dataframe com 3 condições, sendo elas o tipo 1 dark E tipo 2 fire, ou seja, 
# Apenas quando as 3 forem true
Tipo_Poison_Grass = poke.loc[(poke["Type 1"]=="Grass") & (poke["Type 2"]=="Poison") & (poke["Generation"]==1)]
Tipo_Poison_Grass

Tipo_generation = poke.loc[(poke["Type 1"]=="Grass") & (poke["Type 2"]=="Poison") & (poke["Generation"]==1) | (poke["Generation"]==3)]
Tipo_generation

Tipo_lendario = poke.loc[(poke["Legendary"]==True) | (poke["Generation"]==1)]
Tipo_lendario.head(5)

```

**Sorting values** 

Pandas sort_values() function sorts a data frame in Ascending or Descending order of passed Column.

```python
# Embaralhando os dados a partir da coluna Name na ordem crescente que é o padrão
poke.sort_values('Name')

# Embaralhando os dados em ordem decrescente
poke.sort_values('Name', ascending= False)

# Embaralhando duas colunas
poke.sort_values(['Name', 'HP'], ascending = True)

# Embaralhando duas colunas com Type 1 na ordem crescente e HP na ordem decrescente
poke.sort_values(['Type 1', 'HP'], ascending=[1,0])
```

**Adding new collunms**

```python
# Criando uma nova coluna no dataframe poke
poke['Media'] = (poke['HP'] + poke['Attack'] + poke['Defense'])/3
poke.head(5)

# Podemos utilizar a função assign para criar uma coluna com o valor HP original vezes 2
poke2 = poke.assign(HP2 = poke['HP']*2)
poke2.head(5)

# Podemos realizar calculos ao criar novas colunas
poke["Attack%"] = poke['Attack'] * 0.1
poke.head(5)

# Criando uma coluna com uma função que avalia o ataque utilizando if and elses
def AttackValues(value):
    if value < 50:
        return 'Baixo Ataque'
    elif 50 < value < 70:
        return 'Medio Ataque'
    elif 70 < value < 90:
        return 'Bom Ataque'
    else:
         return 'Otimo Ataque'        

poke['goodAttack'] = poke['Attack'].map(AttackValues)
poke         



```

**Filtering Data**

**Vamos aprender metodos para filtrar nossos dados**

```python
# 1º Operadores Lógicos: Podemos usar &, |, strings para fazer nossa seleção
poke[(poke['Attack']> 150)]
poke[(poke['Attack'] > 130) & (poke['Defense'] > 130)]
poke[(poke['Name'] > 'Charmeleon')]

# 2º Utilizado o isin metodo para filtrar apenas os elementos do tipo fogo e water e seus nomes
type = ['Fire', 'Water']
poke[poke['Type 1'].isin(type)]

# Utilizando o isin com números
hp = [130,150]
poke[poke['HP'].isin(hp)]

# Criando um isin com duas condições
poke[(poke['Type 1'].isin(type)) & (poke['HP'].isin(hp))]

#3 Utilizando o Srt acessor

# Escolhendo os pokemons que o nome começam com A
poke[poke["Name"].str.startswith('A')] 

# Escolhendo os pokemons que contém a letra J
poke[poke["Name"].str.contains("J")]

# Escolhendo os pokemons que contém J e com o HP maior que 70
poke[(poke["Name"].str.contains("J")) & (poke["HP"]> 70)]

# Query selector
# poke.query[('poke["Name"] == "Pikachu" & poke["HP"]> 120')]

# # Nlargest
# poke.nlargest(2, poke["HP"]
```

**Utilizando Regex**

```python
import re
#Aqui utilizamos o pacote regex para selecionar os pokemons com tipo 1 igual fogo ou grama, flags = re.I 
#flags = re.I permiti que não tenha problema ser maiusculo ou minisculo
Tipo_Fogo = poke.loc[poke["Type 1"].str.contains("fire|grass", flags = re.I) & (poke["HP"]> 100)]
Tipo_Fogo

# Selecionando pokemons começando com AI
name_ch = poke.loc[(poke['Name']).str.contains("^ch[a-z]*", flags = re.I)]
name_ch
```
